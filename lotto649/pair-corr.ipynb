```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("reshape.csv")
df = pd.DataFrame(np.reshape(df.values, (4088,6)), columns=['1', '2', '3', '4', '5', '6'])
```

```python
pairs = [(1,14), (1,44), (10,31), (10,38), (11,27), (12,19), (13,24), (15,38), (16,19), (17,20), (2,25), (20,39), (20,43), (23,38), (23,47), (25,36), (26,46), (29,33), (3,40), (3,8), (30,43), (32,39), (33,49), (35,48), (4,36), (41,42), (46,48), (5,37), (5,38), (5,45), (5,6), (7,18), (8,12), (8,22), (8,34), (9,21), (9,28)]
```

```python
binary_matrix = np.zeros((len(df), len(pairs)), dtype=int)
```

```python
for i, pair in enumerate(pairs):
    row_contains_pair = ((df == pair[0]) & (df == pair[1])).any(axis=1)
    binary_matrix[:, i] = row_contains_pair.astype(int)

binary_df = pd.DataFrame(binary_matrix, columns=pairs)

binary_df.to_csv('pair_corr.csv')
```
