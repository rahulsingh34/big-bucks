```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.national-lottery.com/canada-6-49/results/"
path = "-archive"
```

```python
for i in range(1982, 2024):
    r = requests.get(url + str(i) + path)
    soup = BeautifulSoup(r.text, 'html.parser')
    results = soup.find('table', class_ = 'table')

    for table in results.find_all('tbody'):
        rows = table.find_all('tr')
        for row in rows:
            try:
                draw = row.find('td', class_ = 'noBefore nowrap').text.strip()
                print(draw)
            except:
                print('skip')
```
